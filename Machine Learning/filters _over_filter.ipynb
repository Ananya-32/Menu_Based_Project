```python
import cv2
import numpy as np

def apply_filters(image):
    # Define laptop screen dimensions (e.g., 1366x768)
    laptop_height, laptop_width = 768, 1366

    # Calculate dimensions for each filtered image
filter_height = laptop_height // 2
filter_width = laptop_width // 3

# Create a blank canvas to place all filtered images
canvas = np.zeros((laptop_height, laptop_width, 3), dtype=np.uint8)

# Resize original image to fit in the grid
image_resized = cv2.resize(image, (filter_width, filter_height))

# 1. Original Image
canvas[0:filter_height, 0:filter_width] = image_resized

# 2. Apply Gaussian Blur
gaussian_blur = cv2.GaussianBlur(image, (15, 15), 0)
gaussian_blur_resized = cv2.resize(gaussian_blur, (filter_width, filter_height))
canvas[0:filter_height, filter_width:filter_width*2] = gaussian_blur_resized

# 3. Apply Median Blur
median_blur = cv2.medianBlur(image, 15)
median_blur_resized = cv2.resize(median_blur, (filter_width, filter_height))
canvas[0:filter_height, filter_width*2:filter_width*3] = median_blur_resized

# 4. Apply Bilateral Filter
bilateral_filter = cv2.bilateralFilter(image, 15, 75, 75)
bilateral_filter_resized = cv2.resize(bilateral_filter, (filter_width, filter_height))
canvas[filter_height:filter_height*2, 0:filter_width] = bilateral_filter_resized

# 5. Apply Edge Detection (Canny)
edges = cv2.Canny(image, 100, 200)
edges_colored = cv2.cvtColor(edges, cv2.COLOR_GRAY2BGR)  # Convert to color image
edges_colored_resized = cv2.resize(edges_colored, (filter_width, filter_height))
canvas[filter_height:filter_height*2, filter_width:filter_width*2] = edges_colored_resized

# 6. Apply Sharpening Filter
kernel = np.array([[0, -1, 0],
                        [-1, 5, -1],
                        [0, -1, 0]])
sharpened = cv2.filter2D(image, -1, kernel)
sharpened_resized = cv2.resize(sharpened, (filter_width, filter_height))
canvas[filter_height:filter_height*2, filter_width*2:filter_width*3] = sharpened_resized

# 7. Apply Emboss Filter
kernel_emboss = np.array([[-2, -1, 0],
                               [-1, 1, 1],
                               [0, 1, 2]])
embossed = cv2.filter2D(image, -1, kernel_emboss)
embossed_resized = cv2.resize(embossed, (filter_width, filter_height))
canvas[filter_height:filter_height*2, filter_width*2:filter_width*3] = embossed_resized
    # Display the combined image with all filters applied
    cv2.imshow('All Filters Applied', canvas)
    cv2.waitKey(0)  # Wait for a key press to close the window

    # Close all windows after all filters are displayed
    cv2.destroyAllWindows()

def capture_image():
    # Initialize webcam
    cap = cv2.VideoCapture(0)
    if not cap.isOpened():
        print("Error: Could not open video capture.")
        return None

    print("Press 'a' to capture an image.")

    while True:
        ret, frame = cap.read()
        if not ret:
            print("Error: Failed to grab frame.")
            break

        cv2.imshow('Webcam', frame)

        # Capture image on 'a' key press
        key = cv2.waitKey(1)
        if key == ord('a'):
            print("Image captured.")
            break

    cap.release()
    cv2.waitKey()
    cv2.destroyAllWindows()
    return frame

if __name__ == "__main__":
    # Capture image from webcam
    captured_image = capture_image()
    if captured_image is not None:
        apply_filters(captured_image)
```
